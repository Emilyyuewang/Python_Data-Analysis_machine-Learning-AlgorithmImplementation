```python
#using location to get the coordinator data
from geopy.geocoders import Nominatim
geolocator=Nominatim()
location=geolocator.geocode("17 5th Avenue NYC")
print(location.address)
print(location.latitude,location.longitude)
```

```python
print(location.raw)
```

```python
#using coordinator data to get the address.
from geopy.geocoders import Nominatim
geolocator=Nominatim()
location=geolocator.reverse("40.79267935 , -73.8409107129498")
print(location.address)
print(location.latitude,location.longitude)
```
